## Deal with Outliers

```python
# Import Package
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# Set up directory
dir_data = 'C:\\Users\\victo\\Dropbox\\AI_M\\data_p1'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# List AMT_ANNUITY quadrant 1 to quadrant 100
q_index = list(range(101))                 
q_all = [np.percentile(app_train[~app_train['AMT_ANNUITY'].isnull()]['AMT_ANNUITY'], q = i) for i in q_index]
print(q_all)
a = pd.DataFrame({'q': q_index,
                  'value': q_all})
a.head()
```

```python
# Use quadrant 50 to impute NAs
print("Before replace NAs, numbers of row that AMT_ANNUITY is NAs: %i" % sum(app_train['AMT_ANNUITY'].isnull()))

q_50 = a.iloc[(51, 1)]
q_50

app_train.loc[app_train['AMT_ANNUITY'].isnull(),'AMT_ANNUITY'] = q_50

print("After replace NAs, numbers of row that AMT_ANNUITY is NAs: %i" % sum(app_train['AMT_ANNUITY'].isnull()))
```

```python
# Normalize AMT_ANNUITY into -1 ~ 1
```

### Hints: Normalize function (to -1 ~ 1)
$ y = 2*(\frac{x - min(x)}{max(x) - min(x)} - 0.5) $

```python
# Self-define normalization function
b = app_train['AMT_ANNUITY']

print("== Original data range ==")
print(b.describe())

def normalize_value(x):
    d1 = ( x - min(x) ) / ( max(x) - min(x) - 0.5)*2
    return d1

app_train['AMT_ANNUITY_NORMALIZED'] = normalize_value(b)

print("== Normalized data range ==")
print(app_train['AMT_ANNUITY_NORMALIZED'].describe())
```

```python
# Use mode to impute AMT_GOOD_PRICE's NAs 
bd = app_train['AMT_GOODS_PRICE']
print("Before replace NAs, numbers of row that AMT_GOODS_PRICE is NAs: %i" % sum(bd.isnull()))

# List the mode of AMT_GOOD_PRICE
from scipy.stats import mode
import time
start_time = time.time()
mode_get = mode(app_train[~bd.isnull()]['AMT_GOODS_PRICE'])
print(mode_get)

value_most = mode_get[0]
print('Mode is %d'% (value_most))

mode_goods_price = list(bd.value_counts().index)
app_train.loc[bd.isnull(), 'AMT_GOODS_PRICE'] = mode_goods_price[0]

print("After replace NAs, numbers of row that AMT_GOODS_PRICE is NAs: %i" % sum(bd.isnull()))
```
