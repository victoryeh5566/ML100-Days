```python
# Import Package
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

## 10 x 10 Matrix Value Between -1 and 1. Visualize with Heatmap

```python
matrix = 2 * np.random.random((10,10)) - 1
plt.figure(figsize=(10,10))
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1, annot = True, vmax = 1)
plt.show()
```

## 1000 x 3 Matrix Value Between -1 and 1. Apply Pairplot to Visualize(scatter, hist and density)

```python
nrow = 1000
ncol = 3
matrix1 = 2 * np.random.random((1000, 3)) - 1
indice = np.random.choice([0,1,2], size = nrow)
plot_data = pd.DataFrame(matrix1, indice)

grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey = False)

# Scatter plot on the top side
grid.map_upper(plt.scatter, alpha = 0.2)
# Histogram on diagonal
grid.map_diag(plt.hist)
# Density plot on the bottom
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)
plt.show()
```

## 1000 x 3 Matrix Value Between -1 and 1. Use Pairplot to Visualize(scatter, hist and density)

```python
nrow = 1000
ncol = 3

matrix2 =  np.random.normal(0, 1, (nrow, ncol))

indice = np.random.choice([0,1,2], size = nrow)
plot_data = pd.DataFrame(matrix2, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey = False)
# Scatter plot on the top side
grid.map_upper(plt.scatter, alpha = 0.2)
# Histogram on diagonal
grid.map_diag(plt.hist)
# Density plot on the bottom
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)
plt.show()
```
