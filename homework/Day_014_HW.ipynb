## Data Prep

```python
# Import package
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# Set up file directory
dir_data = 'C:\\Users\\victo\\Dropbox\\AI_M\\data_p1'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## Visualizae "HOUSETYPE_MODE" and "AMT_CREDIT" Histogram

```python
yy = app_train[['HOUSETYPE_MODE', 'AMT_CREDIT']]
print(yy['HOUSETYPE_MODE'].value_counts())

print(yy.shape)# Identify missing value and Nas
```

```python
## Remove Na
yy = yy.dropna()
yy.head()
```

```python
unique_house_type = np.sort(yy['HOUSETYPE_MODE'].unique())

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    yy.loc[yy['HOUSETYPE_MODE'] == unique_house_type[i], 'AMT_CREDIT'].hist()
    plt.title(str(unique_house_type[i]))
plt.show() 
```
